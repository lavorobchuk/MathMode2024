3. Создайте массив первых 10 нечетных целых чисел.  

```python
ar = np.array([])
count = 0
number = 0

while count < 10:
    if number % 2 != 0:
        ar = np.append(ar, number)
        count += 1
    number += 1

print(ar)
```

4. Создайте массив нулей размера 10 x 10, а затем создайте в нём "рамку" из единиц по краям.  

```python
ar = np.array([[0.]*10]*10)

for i in range(len(ar)):
    for j in range(len(ar[i])):
        if i == 0 or i == 9:
            ar[i][j] = 1
        elif j == 0 or j ==9:
            ar[i][j] = 1
            
print(ar)
```

5. Создайте массив 8 x 8 с шахматной доской из нулей и единиц.  

```python
ar = np.array([[0.]*8]*8)
        
for i in range(len(ar)):
    for j in range(len(ar[i])):
        if i % 2 == 0 and j % 2 == 1:
            ar[i][j] = 1
        if i % 2 == 1 and j % 2 == 0:
            ar[i][j] = 1

print(ar)
```

6. Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def create_array(n):
    ar = np.array([[0.]*n]*n)
    for i in range(len(ar)):
        for j in range(len(ar[i])):
            ar[i][j] = i + j
    return ar
            
print(create_array(8))
```

```python
np.random.seed(1132226486) # мой студенческий билет
print(np.random.randint(10))
```

```python
def f(x):
    return 1/(x**2-2*x-8)

a = 2
b = 3

dx = 0.1

analytical_solution = np.log(abs(b - 4))/6 - np.log(abs(b + 2))/6 - (np.log(abs(a - 4))/6 - np.log(abs(a + 2))/6)

def rectangle_method(f, a, b, dx):
    x_val = np.arange(a + dx / 2, b, dx)
    return np.sum(f(x_val)) * dx

# Метод трапеций
def trapezoid_method(f, a, b, dx):
    x_val = np.arange(a + dx, b, dx)
    return (dx / 2) * (f(a) + 2 * np.sum(f(x_val)) + f(b))

# Метод Симпсона
def simpson_method(f, a, b, dx):
    x_val = np.arange(a + dx / 2, b, dx)
    x_odd_val = x_val[::2]
    x_even_val = x_val[1::2]
    return (dx / 3) * (f(a) + 2 * np.sum(f(x_even_val)) + 4 * np.sum(f(x_odd_val)) + f(b))

tolerance = 1e-5

previous_integral_rectangle = 0
previous_integral_trapezoid = 0
previous_integral_simpson = 0

while True:
    current_integral_rectangle = rectangle_method(f, a, b, dx)
    current_integral_trapezoid = trapezoid_method(f, a, b, dx)
    current_integral_simpson = simpson_method(f, a, b, dx)
    
    difference_rectangle = np.abs(current_integral_rectangle - previous_integral_rectangle)
    difference_trapezoid = np.abs(current_integral_trapezoid - previous_integral_trapezoid)
    difference_simpson = np.abs(current_integral_simpson - previous_integral_simpson)
    
    if (difference_rectangle < tolerance) and (difference_trapezoid < tolerance) and (difference_simpson < tolerance):
        break
    
    previous_integral_rectangle = current_integral_rectangle
    previous_integral_trapezoid = current_integral_trapezoid
    previous_integral_simpson = current_integral_simpson
    
    dx /= 10

print("Аналитическое решение:", analytical_solution)
print("Метод прямоугольников:", current_integral_rectangle)
print("Метод трапеций:       ", current_integral_trapezoid)
print("Метод Симпсона:       ", current_integral_simpson)
```
